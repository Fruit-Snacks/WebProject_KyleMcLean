```python
df = pd.read_csv('DataLocations4.csv', names=['name', 'lat', 'lon', 'pop', 'listing', 'sd', 'mode', 'beds'])
```

```python
df
```

```python
df['text'] = df['name'] + '<br>Average Price ' + (df['pop']).astype(str)
limits = [(0,10),(10,20),(20,30),(30,40),(40,100)]
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)","rgb(255,133,27)","rgb(255,220,0)"]
#color is sorted by number in list

cities = []
scale = 300

for i in range(len(limits)):
    lim = limits[i]
    df_sub = df[lim[0]:lim[1]]
    city = dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df_sub['lon'],
        lat = df_sub['lat'],
        text = df_sub['text'],
        sizemode = 'diameter',
        marker = dict( 
            size = df_sub['pop']/scale, 
            color = colors[i],
            line = dict(width = 2,color = 'black')
        ),
        name = '{0} - {1}'.format(lim[0],lim[1]) )
    cities.append(city)

layout = dict(
        title = '2017 Nashville Renting Price Averages by Area<br>(Click legend to toggle traces)',
        showlegend = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = 'rgb(217, 217, 217)',       
            subunitwidth=1,
            countrywidth=1,
            subunitcolor="rgb(255,65,54)",
            countrycolor="rgb(255, 255, 255)"           
        ),  
    )
    
fig = dict( data=cities, layout=layout )
url = py.plot( fig, validate=False, filename='bubble-map-Nashville-Renting' )
```
